```python
def plot_alles(test):
    ind = 0
    for col in tqdm(test.columns):
        fig = plt.figure()
        plt.hist(test[col])
        #plt.tight_layout()
        plt.savefig(f"plots/plot-{col}.png")
        ind += 1
        plt.close(fig)
```
